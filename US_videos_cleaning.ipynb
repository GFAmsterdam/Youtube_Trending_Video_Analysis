```python
# Import depenencies
import pandas as pd
import json
```

```python
# read in the USvideos.csv file
file = "./Data/USvideos.csv"
us_videos_df = pd.read_csv(file)
us_videos_df.head()
```

```python
# Check the dataset with info() method
us_videos_df.info()
```

```python
# Read in the category file
f = open("./Data/US_category_id.json")
data = json.load(f)
```

```python
# Create a dictionary of category id and description
category_dict = {}

for i in data["items"]:
    category = {i["id"] : i["snippet"]["title"]}
    category_dict.update(category)
```

```python
category_df = pd.DataFrame(category_dict.items(), columns=["category_id", "category_name"])
```

```python
category_df
```

```python
# Save the category_df as csv for future use
category_df.to_csv("./Data/category_id.csv", index=False, index_label=None)
```

```python
# Create a dictionary
categories = {1 : "Film & Animation",
2 : "Autos & Vehicles",
10 : "Music",
15 : "Pets & Animals",
17 : "Sports",
18 : "Short Movies",
19 : "Travel & Events",
20 : "Gaming",
21 : "Videoblogging",
22 : "People & Blogs",
23 : "Comedy",
24 : "Entertainment",
25 : "News & Politics",
26 : "Howto & Style",
27 : "Education",
28 : "Science & Technology",
29 : "Nonprofits & Activism",
30 : "Movies",
31 : "Anime/Animation",
32 : "Action/Adventure",
33 : "Classics",
34 : "Comedy",
35 : "Documentary",
36 : "Drama",
37 : "Family",
38 : "Foreign",
39 : "Horror",
49 : "Sci-Fi/Fantasy",
41 : "Thriller",
42 : "Shorts",
43 : "Shows",
44 : "Trailers"
}
```

```python
# Create a duplicate of 'category_id' to apply the dictionary to
us_videos_df['category_type'] = us_videos_df.loc[:,'category_id']
# Apply the dictionary
us_videos_df['category_type'] = us_videos_df['category_type'].apply(lambda x : categories[x])

# Check the parsed dataframe
us_videos_df.head()
```

```python
# Drop video_id, thumbnail_link, description and the catogery_id columns
us_videos_df = us_videos_df.drop(columns=["video_id", "thumbnail_link", "description"], axis=1)
```

```python
us_videos_df.head()
```

```python
# Convert the trending__date datatype to datetime
us_videos_df["trending_date"] = pd.to_datetime(us_videos_df["trending_date"], format="%y.%d.%m")
```

```python
us_videos_df.head()
```

```python
# Convert the publish_time to datetime
us_videos_df["publish_time"] = pd.to_datetime(us_videos_df["publish_time"], format="%y.%d.%m", exact=False)
```

```python
us_videos_df['publish_time']=us_videos_df['publish_time'].values.astype('datetime64[s]')
```
