```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
youtube_df = pd.read_csv('./USvideos_c1.csv')
youtube_df.dtypes
```

```python
youtube_df
youtube_nums = youtube_df[['views', 'likes', 'dislikes','comment_count']]
```

```python
# Initializing model with K = 3 (since we already know there are three classes of iris plants)
model = KMeans(n_clusters=16, random_state=5)
model
```

```python
# Fitting model
model.fit(youtube_nums)
```

```python
# Get the predictions
predictions = model.predict(youtube_nums)
print(predictions)
```

```python
# Add a new class column to the df_iris
youtube_nums["class"] = model.labels_
youtube_nums.head()
```

```python
youtube_nums.hvplot.scatter(x='comment_count', y='views', by='class')
```
